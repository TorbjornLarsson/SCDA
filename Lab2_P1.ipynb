somewhere between $x=-0.5$ and $x=0$. How can you calculate that extremum more
exactly using the commands you have learned in this lab? Calculate the exact value in the cell below._

```python
# enter your code here

# Calculate the first derivative of p2
derivative2 = p2.deriv(1)

# Find the root(s) of the first derivative to find the x-coordinate of the extremum
x_extremum = derivative2.roots()

# Evaluate p2 at the x-coordinate of the extremum to find the y-coordinate
y_extremum = p2(x_extremum)

print(f"First derivative of p2: {derivative2}")
print(f"X-coordinate of the extremum of p2: {x_extremum[0]}")
print(f"Y-coordinate of the extremum of p2: {y_extremum[0]}")
```
